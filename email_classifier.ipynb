```python
# importing packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
# importing data
spam_df = pd.read_csv('spam.csv')
spam_df
```

```python
spam_df.groupby(by = 'Category').describe()
```

```python
spam_df['spam'] = spam_df['Category'].apply(lambda x : 0 if x =='ham' else  1)
```

```python
spam_df
```

```python
X_train,X_test,y_train,y_test = train_test_split(spam_df['Message'],spam_df['spam'],test_size=0.25)
```

```python
cv = CountVectorizer()
```

```python
X_train_count = cv.fit_transform(X_train)
```

```python
X_train_count.toarray()
```

```python
X_train_count.toarray().shape
```

```python
#pre-test

email = ["hello dude come for the game tonight"]
email_count = cv.transform(email)
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_count,y_train)
```

```python
model.predict(email_count)
```

```python
X_test_count= cv.transform(X_test)
```

```python
model.predict(X_test_count)
```

```python
model.score(X_test_count,y_test)
```

```python
import joblib

joblib.dump(model,'spam_classifier.pkl')
joblib.dump(cv,'vectorizer.pkl')
```
