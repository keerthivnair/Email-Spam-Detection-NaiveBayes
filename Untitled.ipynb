```python
def calculate_prior(df,Y):
    classes = sorted(list(df[Y].unique()))
    prior = []
    for i in classes:
        prior.append((len(df[df[Y] == i]))/len(df[Y]))
    return prior    
```

```python
def calculate_x_given_y(df,Y,feature_name,feature_val):
    df = df[df[Y]==feature_name]

    mean,std = df[feature_name].mean(),df[feature_name].std()

    p_x_given_y = (1 / (np.sqrt(2 * np.pi) * std)) *  np.exp(-((feature_val-mean)**2 / (2 * std**2 )))
    return p_x_given_y
    
```
